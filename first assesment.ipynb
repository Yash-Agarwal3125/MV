```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
from scipy.spatial import distance

def show(img, title="Image", cmap=None):
    plt.figure(figsize=(6, 6))
    if len(img.shape) == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
    
filename='sample.jpg'
```

```python
# 1. OpenCV
img_cv = cv2.imread(filename)
print(f"OpenCV - Type: {type(img_cv)}, Shape: {img_cv.shape} (H, W, Channels)")

# 2. Pillow (PIL)
img_pil = Image.open(filename)
print(f"Pillow - Type: {type(img_pil)}, Size: {img_pil.size} (W, H)")

# 3. Scikit-Image
img_sk = io.imread(filename)
print(f"Scikit - Type: {type(img_sk)}, Shape: {img_sk.shape} (H, W, Channels)")
print("Yash Agarwal 23BAI0011")
```

```python
img_color = cv2.imread(filename)
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)
_, img_binary = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot Binary
axs[0].imshow(img_binary, cmap='gray')
axs[0].set_title('Binary Image')
axs[0].axis('off')

# Plot Grayscale
axs[1].imshow(img_gray, cmap='gray')
axs[1].set_title('Grayscale Image')
axs[1].axis('off')

# Plot Color (Convert BGR to RGB for matplotlib display)
axs[2].imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
axs[2].set_title('Color Image')
axs[2].axis('off')

plt.show()
print("Yash Agarwal 23BAI0011")
```

```python
output_img = img_color.copy()

# 1. Add Text
cv2.putText(output_img, "23BAI0011", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# 2. Add Rectangle (Top-Left, Bottom-Right, Color, Thickness)
cv2.rectangle(output_img, (100, 100), (300, 300), (0, 255, 0), 3)

# 3. Add Circle (Center, Radius, Color, Thickness -1 means filled)
cv2.circle(output_img, (400, 400), 50, (0, 0, 255), -1)

# Display Original vs Output
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
axs[1].set_title('Image with Shapes & Text')
plt.show()
print("Yash Agarwal 23BAI0011")
```

```python
img = cv2.imread(filename)

# Vertical Flip
flip_vert = cv2.flip(img, 0)

# Horizontal Flip
flip_horz = cv2.flip(img, 1)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(cv2.cvtColor(flip_vert, cv2.COLOR_BGR2RGB))
axs[0].set_title('Vertical Flip')
axs[1].imshow(cv2.cvtColor(flip_horz, cv2.COLOR_BGR2RGB))
axs[1].set_title('Horizontal Flip')
plt.show()
print("Yash Agarwal 23BAI0011")
```

```python
img = cv2.imread(filename)
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

# a) Rotate 90 degrees, Scale 0.5 (Downscale)
M_down = cv2.getRotationMatrix2D(center, 90, 0.5)
rotated_down = cv2.warpAffine(img, M_down, (w, h))

# b) Rotate 90 degrees, Scale 2.5 (Upscale)
M_up = cv2.getRotationMatrix2D(center, 90, 2.5)
rotated_up = cv2.warpAffine(img, M_up, (w, h))

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(cv2.cvtColor(rotated_down, cv2.COLOR_BGR2RGB))
axs[0].set_title('Rot 90, Scale 0.5')

axs[1].imshow(cv2.cvtColor(rotated_up, cv2.COLOR_BGR2RGB))
axs[1].set_title('Rot 90, Scale 2.5')
plt.show()
print("Yash Agarwal 23BAI0011")
```

```python
from scipy.spatial import distance
p1 = (50, 50)
p2 = (100, 100)

print(f"Point 1: {p1}, Point 2: {p2}")

# b) Euclidean Distance (L2 Norm) - Straight line distance
dist_euclidean = distance.euclidean(p1, p2)
print(f"Euclidean Distance: {dist_euclidean:.2f}")

# c) City Block Distance (Manhattan / L1 Norm) - Grid path
dist_city = distance.cityblock(p1, p2)
print(f"City Block Distance: {dist_city}")

# d) Chessboard Distance (Chebyshev / L-infinity) - Max of differences
dist_chess = distance.chebyshev(p1, p2)
print(f"Chessboard Distance: {dist_chess}")

# a) Distance Transform (OpenCV Specific)
dist_transform = cv2.distanceTransform(img_binary, cv2.DIST_L2, 5)

plt.figure(figsize=(6,6))
plt.imshow(dist_transform, cmap='gray')
plt.title("Distance Transform (Visualized)")
plt.axis('off')
plt.show()
print("Yash Agarwal 23BAI0011")
```
